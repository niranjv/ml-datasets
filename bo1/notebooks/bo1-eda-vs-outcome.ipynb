```python
# Imports

# embed static images in notebook
%matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina'


import platform
print('python.version = ', platform.python_version())
import IPython
print('ipython.version =', IPython.version_info)

import numpy as np
print('numpy.version = ', np.__version__)
import pandas as pd
print('pandas.version = ', pd.__version__)

import sklearn
print('sklearn.version = ', sklearn.__version__)

import matplotlib as mpl
print('matplotlib.version = ', mpl.__version__)
import matplotlib.pyplot as plt
import seaborn as sns
print('seaborn.version = ', sns.__version__)
```

```python
df = pd.read_csv("../data/processed/train.csv", index_col=0)

df["device_type"] = df["device_type"].astype('category')
df["gender"] = df["gender"].astype('category')
df["in_initial_launch_location"] = df["in_initial_launch_location"].astype('category')
df["outcome"] = df["outcome"].astype('category')
```

## Explore categorical covariates
---

```python
df.describe(exclude=[np.number])
```

```python
sns.factorplot(x="device_type", col="outcome", data=df, kind="count");
```

```python
sns.factorplot(x="gender", col="outcome", data=df, kind="count");
```

```python
sns.factorplot(x="in_initial_launch_location", col="outcome", data=df, kind="count");
```
