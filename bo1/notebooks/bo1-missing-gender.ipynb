```python
df = pd.read_csv("../data/processed/train.csv", index_col=0)

df["device_type"] = df["device_type"].astype('category')
df["gender"] = df["gender"].astype('category')
df["in_initial_launch_location"] = df["in_initial_launch_location"].astype('category')
df["outcome"] = df["outcome"].astype('category')
```

```python
idx_missing_gender = pd.isnull(df["gender"])
missing_gender = df[idx_missing_gender]
missing_gender.shape
```
